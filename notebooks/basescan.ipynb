```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
import random
# Initialize the WebDriver (e.g., using Chrome)
driver = webdriver.Chrome()

# Open the website
driver.get("https://basescan.org/txs?ps=100&p=356")

# Wait for user to solve the CAPTCHA manually
input("Please solve the CAPTCHA and then press Enter to continue...")

# Function to scrape the current page
def scrape_current_page():
    rows = driver.find_elements(By.CSS_SELECTOR, "table tbody tr")
    page_data = []
    for row in rows:
        cols = row.find_elements(By.TAG_NAME, "td")
        page_data.append([col.text for col in cols])
    return page_data

# Scrape the first page
data = scrape_current_page()

# Scrape the next 49 pages
for _ in range(1, 500):  # Adjust the range if you need more or fewer pages
    try:
        # Find and click the "Next" button
        print("page: ", _)
        next_button = driver.find_element(By.CSS_SELECTOR, 'a.page-link[aria-label="Next"]')
        next_button.click()
        
        # Wait for the next page to load
        time.sleep(1)
        
        # Scrape the current page
        data.extend(scrape_current_page())

        
    except Exception as e:
        print(f"An error occurred: {e}")
        break

# Close the WebDriver
driver.quit()

# Check the first row to see the number of columns
if data:
    num_columns = len(data[0])
    print(f"Number of columns in the first row: {num_columns}")

# Adjust the columns list based on the number of columns
columns = ["Txn Hash", "Method", "Block", "Age", "From", "To", "Value", "Txn Fee"]
if num_columns != len(columns):
    print("Adjusting columns to match the scraped data")
    columns = [f"Column {i+1}" for i in range(num_columns)]

# Convert to DataFrame
df = pd.DataFrame(data, columns=columns)

# Save to CSV
df.to_csv("transactions.csv", index=False)

print("Data has been scraped and saved to transactions.csv")
```
